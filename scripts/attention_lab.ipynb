```python
# Phase 1: Installing libraries
!pip install -q transformers torch bertviz

# Phase 2: Loading the Portuguese BERTimbau model
import torch
from transformers import AutoTokenizer, BertModel
from bertviz import head_view

model_name = "neuralmind/bert-base-portuguese-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name, output_attentions=True)

# Phase 3: Tokenizing the target phrase
text = "Deita-se o preparado numa forma em forma de família."
inputs = tokenizer(text, return_tensors='pt')
tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])

# Phase 4: Running model inference
with torch.no_grad():
    outputs = model(**inputs)
attention = outputs.attentions

# Phase 5: Generating the interactive attention map
print("Generating attention map...")
head_view(attention, tokens, html_action='view')
```
