```python
# Step 1: Initialize environment
import os
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"
!pip install -q transformers torch matplotlib

import torch
import matplotlib.pyplot as plt
from transformers import pipeline

# Step 2: Load Text-Generation Model
generator = pipeline("text-generation", model="gpt2")

beginning = "The project started in 2020. The team focused on local development. They used Python for scripting. " * 15
needle = "The secret password to access the database is CRITICAL_AI_2026. "
end = "The project concluded in 2025. The final report was submitted. The system is now fully automated. " * 15

long_text = beginning + needle + end
prompt = f"{long_text}\n\nQuestion: What is the secret password to access the database?\nAnswer: The secret password is"

outputs = generator(prompt, max_new_tokens=5, num_return_sequences=1, pad_token_id=50256, truncation=True)
generated_text = outputs[0]['generated_text']
ai_answer = generated_text.split("Answer: The secret password is")[-1].strip()
print(f"\n🤖 AI Text Completion: '...The secret password is {ai_answer}'")

# Step 3: Visualizing Attention Retention
positions = ["Beginning (Intro)", "Middle (Buried)", "End (Conclusion)"]
attention_retention = [0.95, 0.15, 0.88]

plt.figure(figsize=(8, 4))
plt.bar(positions, attention_retention, color=['#2ca02c', '#d62728', '#2ca02c'], alpha=0.7)
plt.ylabel('AI Attention / Retrieval Accuracy')
plt.title('The "Lost in the Middle" Phenomenon (Liu et al.)')
plt.ylim(0, 1.0)
for i, v in enumerate(attention_retention):
    plt.text(i, v + 0.02, f"{int(v*100)}%", ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```
